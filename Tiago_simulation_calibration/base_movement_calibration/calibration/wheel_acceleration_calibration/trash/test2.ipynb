```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range of x values
x = np.linspace(0, 2, 400)
x_exp = np.linspace(-1, 2, 400)  # For functions with broader domains

# Define the functions to be plotted
functions = {
    "Quadratic": lambda x: x**2,
    "Cubic (x >= 0)": lambda x: x**3,
    "Exponential": lambda x: np.exp(x),
    "Negative Cosine (0, pi)": lambda x: -np.cos(x),
    "Hyperbolic Sine": lambda x: np.sinh(x),
    "Hyperbolic Cosine": lambda x: np.cosh(x),
    "x * ln(x) (x >= e)": lambda x: x * np.log(x),
    "e^(x^2)": lambda x: np.exp(x**2),
    "1 / (1 - x)": lambda x: 1 / (1 - x),
}

# Create the plot
plt.figure(figsize=(12, 8))

for label, func in functions.items():
    if label == "x * ln(x) (x >= e)":
        # Avoid negative values for logarithmic function
        x_vals = np.linspace(np.exp(1), 2, 400)
        plt.plot(x_vals, func(x_vals), label=label)
    else:
        plt.plot(x, func(x), label=label)

# Customize plot appearance
plt.title('Monotonically Increasing and Convex Functions')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.ylim([-2, 10])
plt.xlim([0, 2])
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.legend(loc="upper left")
plt.grid(True)

# Display the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Assign specific values for a, k, and B
a_value = 2
k_value = 3
B_value = -5

# Define the f(x) and g(x) functions based on the solved values for S and S'
def f_function(x, a, k, B):
    S = -B / a**2
    S_prime = (2 * B + a * k) / a
    return S * x**2 + S_prime * x

def g_function(x, k, B):
    return k * x + B

# Generate x values and compute f(x) and g(x)
x_values = np.linspace(-5, 5, 400)
f_values = f_function(x_values, a_value, k_value, B_value)
g_values = g_function(x_values, k_value, B_value)

# Plot f(x) and g(x)
plt.figure(figsize=(8, 6))
plt.plot(x_values, f_values, label=r'$f(x) = Sx^2 + S^\prime x$', color='blue')
plt.plot(x_values, g_values, label=r'$g(x) = kx + B$', color='red', linestyle='--')
plt.axvline(x=a_value, color='gray', linestyle=':')
plt.title('Comparison of f(x) and g(x)')
plt.xlabel('x')
plt.ylabel('Function values')
plt.legend()
plt.xlim(0, 10)
plt.ylim(0, 20)
plt.grid(True)
plt.show()
```
